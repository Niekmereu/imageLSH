```python
from ImageFinderLSH import ImageFinder, hamming_similarity
import matplotlib.pyplot as plt
```

```python
HASH_SIZE = 200
N_BANDS = 5
THRESHOLD = 0.7
```

```python
model = ImageFinder(hash_size=HASH_SIZE, n_bands=N_BANDS)
model.process_input_dir("images/")
similar_lsh = model.find_similar_img("images/ambipom.png", threshold=THRESHOLD)
similar_lsh
```

```python
model._hashed_buckets
```

```python
accuracy_list = []
for n in range(1, 20):
    model = ImageFinder(hash_size=HASH_SIZE, n_bands=n)
    model.process_input_dir("images/")
    similar_lsh = model.find_similar_img("images/aegislash-blade.png", threshold=THRESHOLD)
    similar_true = []
    for sgn in model.signatures:
        similarity = hamming_similarity(model.image_signature, model.signatures[sgn], HASH_SIZE)
        if similarity > THRESHOLD:
            similar_true.append((sgn, similarity))
    accuracy = len(similar_lsh)/len(similar_true)
    accuracy_list.append(accuracy)
```

```python
fig = plt.plot(range(1, 20), accuracy_list)
plt.xlabel("Number of Bands")
plt.ylabel("Accuracy")
```
